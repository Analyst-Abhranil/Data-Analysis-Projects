```python
data2=pd.read_csv('deliveries.csv')
data2.head(10)
```

```python

data2.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# against which opponent did VK score highest run?
```

```python
def get_opposition(batsman):
    mask=data2['batsman']==batsman
    return data2[mask].groupby('bowling_team')[['bowling_team','batsman_runs']].sum().sort_values('batsman_runs',ascending=False).index[0]
```

```python
 get_opposition('V Kohli')
```

```python
 get_opposition('SR Tendulkar')
```

```python
 get_opposition('G Gambhir')
```

```python
# Find out top 5 most destructive death over batsmen
# 16-20 overs-death, minimum 200 balls played
```

```python
death_over=data2[data2['over']>15]
death_over.head()
```

```python
balls_played=death_over.groupby('batsman')['batsman_runs'].count()
balls_played
```

```python
eligible_batsmen=balls_played[balls_played>200]
eligible_batsmen
```

```python
eligible_batsmen_list=eligible_batsmen.index.tolist()
eligible_batsmen_list
```

```python
final=death_over[death_over['batsman'].isin(eligible_batsmen_list)]
final.head()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
runs=final.groupby('batsman')['batsman_runs'].sum()
sr=(runs/balls)*100
sr
```

```python
sr.sort_values(ascending=False).head()
```
